```python
import requests as r
import pandas as pd
from bs4 import BeautifulSoup as bs

def get_emissions(base_url, makers):
    emissions = []
    for maker in makers:
        page = 1
        while True:
            make_url = f"{base_url}PowerSearch.do?action=noform&path=4&year1=2023&year2=2025&make={maker}&srchtyp=newMake&pageno={page}&rowLimit=50"
            response = r.get(make_url)
            soup = bs(response.content, 'html.parser')
            links = [x.get("href") for x in soup.select("a[href]") if "yourMpgVehicle" not in x.get("href")]

            if not links:
                break

            for link in links:
                response = r.get(base_url + link)
                soup = bs(response.content, 'html.parser')
                type = soup.find("td", class_="fuel nobottomborder padding").getText(strip=True)

                if type == "Electricity":
                    emissions.append(soup.find_all("td", class_="even sbsCellData")[2].getText().split(" ")[0]) # hrs per 240v
    print("Creating DataFrame.")
    emission_dict = {"emissions": emissions}
    return pd.DataFrame(emission_dict)

base_url = "https://www.fueleconomy.gov/feg/"
makers = ["Chevrolet"]  # Example: You can include multiple manufacturers here
manufacturer_df = get_emissions(base_url, makers)
print(manufacturer_df)
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Main URL
main_url = "https://www.fueleconomy.gov/feg/"
response = requests.get(main_url)
soup = BeautifulSoup(response.content, "html.parser")

url = "https://www.fueleconomy.gov/feg/PowerSearch.do?action=noform&year1=2019&year2=2024&cbmkChevrolet=Chevrolet&cbmkFord=Ford&cbmkNissan=Nissan&minmsrpsel=0&maxmsrpsel=0&city=0&hwy=0&comb=0&YearSel=2019-2024&make=Chevrolet%3B+Ford%3B+Nissan&mclass=&vfuel=&vtype=&trany=&drive=&cyl=&MpgSel=000&sortBy=Comb&Units=&url=SearchServlet&opt=new&minmsrp=0&maxmsrp=0&minmpg=0&maxmpg=0&sCharge=&tCharge=&startstop=&cylDeact=&rowLimit=200&pageno=1&tabView=1"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

emissions = []

# Find all emission components
components = soup.find_all("div", class_="ghg-score")

for component in components:
    # Extract emission text
    emission_text = component.get_text()
    # Check if the emission text contains 'grams'
    if 'grams' in emission_text:
        # Split and extract the numeric part
        emission_value = float(emission_text.split()[0])
        # Append to emissions list
        emissions.append(emission_value)

# Create DataFrame
df = pd.DataFrame({'emission': emissions})
df
```
